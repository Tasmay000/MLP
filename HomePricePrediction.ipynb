```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings("ignore")
```

```python
trainset = pd.read_excel("Rent TrainSingle.xlsx")
trainset
```

```python
plt.xlabel = ("area")
plt.ylabel  = ("rent")
plt.scatter(trainset.area,trainset.rent,marker = '+',color = 'red')
plt.show()
```

```python
reg = LinearRegression()
reg.fit(trainset[['area']],trainset.rent)
```

```python
reg.predict([[2000]])
```

```python
testset = pd.read_excel("Area TestSingle.xlsx")
testset
```

```python
p = reg.predict(testset)
```

```python
testset['prices'] = p
testset.head(5)
```
